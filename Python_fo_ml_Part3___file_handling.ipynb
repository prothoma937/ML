1. Practice :


* get the txt from sample.txt and count the folowing
* number of lines
* njmber of words
* number of characters
* and save it to the counter of string.txt file


```python
from functools import reduce
with open ("./sample_data/sample.txt","r") as file:
  string=file.readlines()
  total_lines=len(string)
  print(string)
  print(total_lines)
  #word
  number_words=list(map(lambda x: len(x.split()),string))
  print(number_words)
  total_number_words=reduce(lambda x,y: x+y,number_words)
  print(total_number_words)

  #clean remoce lewline
  string=list(map(str.strip,string))
  print(string)
  #remove sapace
  string=list(map(lambda x:x.replace(" ",""),string))
  number_char=list(map(lambda x: len(x),string))
  print(string)
  print(number_char)
  total_number_char=reduce(lambda x,y: x+y,number_char)
  print(total_number_char)

with open ("./sample_data/counter_of_string.txt","w") as file:
  file.write(f"total line:{total_lines}\ntotal num of word:{total_number_words}\ntotal number of charcter:{total_number_char}")
```
