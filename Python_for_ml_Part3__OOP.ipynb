#OOP Inheritance

```python
#single and multiple(attribute inhherit and mrthod)
class Smartphone(Phone):
  def __init__(self,model,battery,camera,processor):
   super().__init__(model,battery,camera)
   self.processor=processor


  def charge(self,hour):
    print("first charge in process")
    super().charge(hour)

vivo = Samartphone("vivos2", 500, 40, "Snapdragon")
print(vivo.model)
vivo.charge(5)


class Cooling:
  def __init__(self,cooling):
    self.cooling=cooling
#method
  def cooling_on(self):
    print(f"the system id being cool by {self.cooling}" )
```

```python
from typing import Self
#multiple inheritance
class Smartphone_coolingmode(Smartphone,Cooling):
 def __init__(self,model,battery,camera,processor,cooling_method):
  Smartphone.__init__(self,model,battery,camera,processor)
  Cooling.__init__(self,cooling_method)
  self.cooling_method = cooling_method
```

```python
pro_cooling=Smartphone_coolingmode("Y",500,110,"sd","nitrogen")
print(pro_cooling.processor)
print(pro_cooling.battery)
print(pro_cooling.cooling_method)
print(pro_cooling.cooling_on())
print(pro_cooling.charge(1))
```

#polymorphism

```python
#method overwriting
class Camera:
    def __init__(self, name):
        self.name = name

    def capture(self):
        print(f"{self.name} is capturing a photo!")
```

```python

class Smart(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

#overwriting
    def capture(self):
      print("photo is captured by phone")

class Dsr(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution
    #overwriting
    def capture(self):
      print("photo is captured by dslr")

class Done(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution
    #overwriting
    def capture(self):
      print("photo is captured by done")


phone = Smart("Phone", 49)
dslr = Dsr("dslr", 44)
done = Done("done", 30)

print(phone.name, phone.resolution)
print(dslr.name, dslr.resolution)
print(done.name, done.resolution)

phone.capture()
dslr.capture()
done.capture()
```
